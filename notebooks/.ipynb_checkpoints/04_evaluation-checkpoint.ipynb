# Traffic Sign Recognition - Evaluation
In this notebook, we will evaluate our simulated CNN model.  
We'll visualize predictions, analyze performance with a confusion matrix, and discuss which classes are often confused.

## Step 1 - Imports 


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
```

# Step 2 - Simulated Predictions
Since we don't have a real trained model, we'll reuse the predictions from 03_model_training.ipynb.  
In practice, you'd load your trained model and predict on the test set.


```python
import numpy as np

# Simulate test set (same as in model training notebook)
np.random.seed(42)
num_samples = 20       # number of test images
img_height, img_width, channels = 32, 32, 3
num_classes = 5        # number of classes

X_test = np.random.rand(num_samples, img_height, img_width, channels)
y_test = np.random.randint(0, num_classes, size=(num_samples,))

y_test_pred_prob = np.random.rand(len(y_test), num_classes)

# Convert probabilities to predicted class labels
y_test_pred = np.argmax(y_test_pred_prob, axis=1)


print("Simulated test set ready. Number of samples:", len(y_test))
print("First 5 true labels:", y_test[:5])

```

## Step 3 - Accuracy
Accuracy is the simplest metric: the percentage of correctly classified samples.  


```python
acc = accuracy_score(y_test, y_test_pred)
print(f"Simulated Test Accuracy: {acc*100:.2f}%")
```

## Step 4 - Confusion Matrix 
A **confusion matrix** shows how well the model predicts each class:  

- Rows = true labels  
- Columns = predicted labels  
- Diagonal = correct predictions  
- Off-diagonal = misclassifications (“confused” classes)

```python
cm = confusion_matrix(y_test, y_test_pred, labels=np.arange(num_classes))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(num_classes))
disp.plot(cmap=plt.cm.Blues)
plt.title("Simulated Confusion Matrix")
plt.show()
```

From the confusion matrix, we can see which classes are often mistaken for others.  
For example, class 2 might be confused with class 3 if their visual patterns are similar.


## Step 5 - Visualizing Example Predictions
We can also look at a few images and see their predicted vs true labels.
- **T:** True label  
- **P:** Predicted label  

Even in a real dataset, this helps to visually inspect **common mistakes**.

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(X_test[i])
    ax.set_title(f"T:{y_test[i]} P:{y_test_pred[i]}")
    ax.axis('off')
plt.suptitle("Example Predictions")
plt.show()
```

## Step 6 - Key Takeaways
- **Accuracy** gives an overall measure of performance.  
- **Confusion matrix** highlights which classes are “confused” by the model.  
- Visual inspection of predictions can help understand why certain mistakes happen.  
- In a real project, you would improve performance by:
  - Using more training data
  - Data augmentation
  - Fine-tuning model architecture
  - Hyperparameter optimization
