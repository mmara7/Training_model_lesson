# Traffic Sign Recognition – Data Exploration

In this notebook, we explore the traffic sign dataset and understand
its structure before training a machine learning model.

We focus on understanding the problem and fixing it step by step.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Understanding the dataset

The dataset contains images of traffic signs along with their labels.
Each image represents a specific traffic sign category.

Before training a neural network, we need to:
- Inspect the data shape
- Visualize sample images
- Understand pixel values

```python
# Simulated dataset: 100 images, 32x32 pixels, 3 color channels
images = np.random.randint(
    0, 256, size=(100, 32, 32, 3), dtype=np.uint8
)

# Simulated labels (43 traffic sign classes)
labels = np.random.randint(0, 43, size=(100,))
```

```python
print("Images shape:", images.shape)
print("Labels shape:", labels.shape)
```

## Step 2: Visualizing sample images

Visualizing images helps us understand:
- Image quality
- Colors and lighting
- Differences between classes

```python
plt.figure(figsize=(8, 4))

for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i])
    plt.axis("off")
    plt.title(f"Label: {labels[i]}")

plt.show()
```

## Step 3: Pixel value distribution

Images are stored as pixel values between 0 and 255.
Neural networks perform better when inputs are normalized.

```python
normalized_images = images / 255.0

print("Min pixel value:", normalized_images.min())
print("Max pixel value:", normalized_images.max())
```

### Why this helps

Neural networks work with mathematical operations.
Smaller, consistent input values make these operations easier and more reliable.

This simple step often makes a bigger difference than changing the model itself.


## Step 4: Image size consistency

Convolutional Neural Networks require all input images
to have the same shape.

In real-world datasets, images may have different sizes.
A common solution is resizing all images to a fixed resolution
(e.g. 32×32 pixels).

In this walkthrough, images are already consistent in size,
so no resizing is required at this stage.

```python
print("Final dataset shape:", normalized_images.shape)
```
